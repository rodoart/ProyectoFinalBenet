```julia
"""
initial_conditions(rango_re, rango_im)

Esta función crea una matriz de números complejos en el rango especificado (rango_re x rango_im).

"""


function initial_conditions(rango_re, rango_im)
    
    imaginarios = collect(rango_im)
    reales = collect(rango_re)
    
    Z_initial = complex( zeros( length(rango_re), length(rango_im) ))
    
    for j in 1 : length(rango_im)
        for i in 1 : length(rango_re)
            
            Z_initial[i,j] = complex(reales[i], imaginarios[j])
            
        end
    end
    
    Z_initial
    
end
```

```julia
"""
    iterator(f, X0, n)

Esta función itera n veces todos los valores del arreglo X0 con una función f.
"""
function iterator(f, X0, n)
    
    for i in 1:length(X0[:,1])
        for j in 1:length(X0[1,:])
            for t in 1:n

                X0[i,j] = f(X0[i,j])
                

            end
        end
    end

    X0
    
end
```

Usando un artículo del portal [Linux Voice](https://www.linuxvoice.com/issues/010/julia.pdf), nos pudimos orientar en como poder graficar las iteracciones de la función que nos dieron en el ejercicio.

```julia
"""
    julia_fractal(f, X0, n, k)
    OUT: matriz_iteraciones

La función julia_fractal itera de `1` hasta `n`, una función `f` dada. Sin embargo, lo que hace esta función es
marcar hasta que número de iteración la función diverge, para una condición inicial. El parámentro k es el que 
establece este límite, los números cuya norma sea mayor a `k`, divergen. 
"""

function julia_fractal(f, X0, n, k)
    
    num_iteraciones = zeros(length(X0[:,1]), length(X0[1,:])) #Arreglo de mismas dimensiones que el arreglo complejo                                                              #jo 
    #X0, pero este es entero.
    
    for i in 1:length(X0[:,1]) #Recorre cada elemento de la matriz de condiciones iniciales.
        for j in 1:length(X0[1,:])
            for t in 1:n     # número de iteraciones

                X0[i,j] = f(X0[i,j]) #itera la función 
                num_iteraciones[i,j] = t # por cada iteración, se actaliza el valor de las iteraciones.
                
                if abs(X0[i,j]) >= k  #si una entrada de X0 es mayor a k, entonces esta diverge. El k mínimo para que
                    #salga algo es 4.
                    
                    break
                    
                end 

            end
        end
    end
   
    
    num_iteraciones
    
end
```

```julia
Z0 = initial_conditions(-1:1/2^8:1, -1:1/2^8:1)
Zf = julia_fractal(f, Z0, 100, 4) #el valor de k es 4, y el número de iteraciones es 100.
```

```julia
imshow(Zf)
title(L"f(z) = z^2 -0.123 +0.745i ")
```

**Observaciones:** La clave está en graficar el número de iteraciones para una condición inicial antes de que diverja.

```julia
"""
    compute_roots(f, f´, range)

La función encuentra las raices de una función dada usando el método de Newton. `f` es la función a la que se quiere
encontrar las raíces, `f'` es la derivada de esa función, y `range` es el rango de condiciones iniciales donde se 
aplica el método de Newton. 
"""


function compute_roots(f, f´, range)
    
    initial_conditions = range
    roots = similar(initial_conditions)

    for (j, x_0) in enumerate(initial_conditions)
        x = x_0

        # 100 iterations of Newton's method
        for i in 1:100
            x = x - f(x) / f´(x)
        end

        roots[j] = x
    end
    
    roots
end
```

```julia
g(x) = 2x%1 #Mapeo
#g(x) = 2*x < 1 ? 2*x : 2*x-1  #Esta función la escribió Luis, pero no le entiendo del todo.
dg(x) = 2
```

Definiendo un arreglo de condiciones iniciales:

```julia
mirango = -1:0.01:1
groots = compute_roots(g, dg, -1:0.01:1)
plot(mirango, groots)
ylim(-1.1, 1.1)
xlim(-1.1, 1.1)

xlabel("Condiciones iniciales")
ylabel(L"Raíces de función $g(x)$")
title(L"Cálculo de las raíces de la función $g(x)$")
```

**Observaciones:** 

> Para valores iniciales en el intervalo $[-1,1]$, las raíces de $g(x) = 2x \ mod \ 1$ son: $-0.5$, $0$ y $0.5$.

> Ahora queremos saber cuales de esos puntos cumplen con la condición de punto fijo, es decir, que se mapeen en sí mismos:

```julia
g(0.0) #sí es punto fijo.
```

```julia
g(0.5) #No es punto fijo.
```

```julia
g(-0.5) #Tampoco es punto fijo.
```

> Por tanto, el único punto fijo del sistema es el $0$.

```julia
"""
    mapeos(f, X0, n)

La función toma un arreglo de una dimensión de condiciones iniciales `X0`, e itera una función `f` para todos las 
entradas de `X0`. `n`es el número de iteraciones.
"""

function mapeos(f, X0, n) #Estoy considerando a X0 como un array de 1 x n dimensiones.
    
    x_solution = zeros(eltype(X0), length(X0), n)   #La matriz final es contiene las orbitas para todas las cond. iniciales.
    x_solution[:, 1] = X0       #Meto condiciones iniciales a la matriz de soluciones.
    
    for t in 2:n
        for i in 1:length(X0)
        
            x_solution[i,t] = f(x_solution[i, t-1]) #iteración
            
        end
        
    end
    
    x_solution
    
end
```

```julia
t = collect(1:100)
#X0 = collect(0.1:0.1:1)
X0=[2//5, 1//27, 1//3, 1//7]
Xf = mapeos(g, X0, 100)

figure(figsize=(12,4))

for i in 1:length(Xf[:,1])
    etiqueta = Xf[i,1]
    plot(t, Xf[i,:]', label=("x0 = $etiqueta"))
    #plot(t, Xf[i,:]', ".")
    
end

#plot(t, Xf[5,:]', label=("x0 = 0.5"))
#plot(Xf[:,:], label=("x0 = 0.5"))

ylim(-0.1, 1.1)
xlabel("Iteraciones")
ylabel("g(x)")
title("Mapeos de función g(x) con condiciones iniciales en el intervalo [0,1]")
legend()
```

```julia
Xf #Comentario de Luis: Checar que con números racionales como condiciones iniciales, las orbitas no converjen a cero 
#cuando n>64.
```

```julia
X2 = collect(0:0.3:1)
t = collect(1:70)
orbitas = mapeos(g, X2, 70)

figure(figsize=(12,4))

for i in 1:length(orbitas[:,1])
    etiqueta = orbitas[i,1]
    plot(t, orbitas[i,:]', label=("x0 = $etiqueta"))
    plot(t, orbitas[i,:]', ".")
    
end

ylim(-0.1, 1.1)
xlabel("Iteraciones")
ylabel(L"g(x)")
title(L"Mapeos de función $g(x)$ con condiciones iniciales en el intervalo $[0,1]$")
legend()
```

**Observaciones:** Se puede apreciar con las dos últimas gráficas, que cuando se usan números racionales las órbitas continúan pasando la iteración $55$, sin embargo, cuando se usan condiciones iniciales que son tipo `float`, las orbitas convergen a cero después de la iteración $50$. Esto se debe a la manera en que la computadora saca el módulo 1 de un número tipo float. Todo número flotante con precisión 64 se convierte a un binario con 64 cifras de precisión. Cuando se saca el módulo de un número flotante, la computadora recorre un lugar el punto decimal en su representación binaria. Cuando llega a la iterada 64 las cifras se acaban, entonces el siquiente número que la computadora pondrá después del punto decimal es un cero. Por eso se mapean a cero cuando n>64.

**Estabilidad de puntos fijos**

```julia
X2 = [0, -1//2, 1//2]
t = collect(1:70)
orbitas = mapeos(g, X2, 70)

figure(figsize=(12,4))

for i in 1:length(orbitas[:,1])
    etiqueta = orbitas[i,1]
    plot(t, orbitas[i,:]', label=("x0 = $etiqueta"))
    plot(t, orbitas[i,:]', ".")
    
end

ylim(-1.1, 1.1)
xlabel("Iteraciones")
ylabel(L"g(x)")
title(L"Mapeos de función $g(x)$ evaluados en puntos fijos")
legend()
```

El único punto fijo que se mapea a sí mismo es $0$, es decir que este es un punto fijo estable. Los puntos $0.5$ y $-0.5$ convergen a cero, es decir, no cumplen con la condición para ser puntos fijos.
