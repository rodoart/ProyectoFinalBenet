## Sección 2:"Ramping up"

### Ejercicio 3
```
git checkout C6               # Mueve a HEAD al commit marcado como C6
git branch -f master HEAD     # Mueve a la rama de master a donde se encuentra HEAD(C6)
git branch -f bugFix HEAD~4   # Mueve a la rama de bugFix cuatro comits anteriores a HEAD(C0)
git checkout HEAD~3           # Mueve a HEAD a 3 comits anteriores(C1)
```

![alt text](ImagenesT1/s2e3.png "Sección Ramping up, Ejercicio 3")

### Ejercicio 4
```
git reset HEAD^        # Regresa a HEAD(rama local) un commit anterior
git checkout pushed    # Coloca a HEAD en la rama pushed
git revert HEAD        # Regresa "remotamente" a pushed un commit atrás y lo salva
```

![alt text](ImagenesT1/s2e4.png "Sección Ramping up, Ejercicio 4")

## Sección 4: "A mixed bag"

### Ejercicio 1
```
git rebase -i HEAD~3        #Coloca a bugFix tres commits atrás(C1), abre la ventana de diálogo para selccionar C4
git branch -f master HEAD   #Mueve la rama master a donde se encuentra HEAD(C4', junto con bugFix)
```

![alt text](ImagenesT1/s4e1.png "Sección A mixed bag, Ejercicio 1")

### Ejercicio 3
```
git checkout master            #Coloca al marcador en master
git cherry-pick C2             #Agarra al commit C2 y lo coloca en outra rama con master como C2'
git branch -f master HEAD^     #Regresa a master a su posición original (C1)
git cherry-pick C2' C3         #Agarra a los commits C2' y C3 (reordenamiento) y coloca a master al final de la rama
```

![alt text](ImagenesT1/s4e3.png "Sección A mixed bag, Ejercicio 3")

### Ejercicio 4
